Dropping columns with very high correlation

```python
corr_matrix = df.corr().abs()
```

```python
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape),k=1).astype(np.bool)) # selecting the upper triangle of corr_ matrix
```

```python
to_drop = [column for column in upper.columns if any(upper[column] > 0.8)]
```

Advertising data

```python
df1=pd.read_csv('Advertising.csv')
```

```python
df1=df1[['TV','radio','newspaper','sales']]
df1.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lg1=LinearRegression()
lg2=LinearRegression()
lg3=LinearRegression()
lg0=LinearRegression()
```

```python
m1=lg1.fit(df1[['TV']],df1['sales'])
m2=lg2.fit(df1[['radio']],df1['sales'])
m3=lg3.fit(df1[['newspaper']],df1['sales'])
m0=lg0.fit(df1[['TV','radio','newspaper']],df1['sales'])
```

```python
print(m1.coef_)
print(m2.coef_)
print(m3.coef_)
print(m0.coef_)
```

```python
import matplotlib.pyplot as plt

corr = df1.corr()
corr.style.background_gradient(cmap='coolwarm').set_precision(3)
```
